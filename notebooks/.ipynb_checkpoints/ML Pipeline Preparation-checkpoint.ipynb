# ML Pipeline Preparation
Follow the instructions below to help you create your ML pipeline.
### 1. Import libraries and load data from database.
- Import Python libraries
- Load dataset from database with [`read_sql_table`](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_sql_table.html)
- Define feature and target variables X and Y

```python
# import libraries
import nltk
nltk.download(['punkt', 'wordnet', 'averaged_perceptron_tagger', 'stopwords'])

import re
import numpy as np
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

from sqlalchemy.orm import sessionmaker
from sqlalchemy import create_engine, MetaData,  select, func, Integer, Table, Column

from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn import multioutput
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
import pickle

from sklearn.metrics import precision_recall_fscore_support
```

```python
# load data from database
table_name = 'DisasterResponseData'
engine = create_engine('sqlite:///../data/DisasterResponse.db')
df = pd.read_sql("SELECT * FROM " + table_name, engine)
# Read columns from dataframe into arrays for ML
X = df['message']
y = df.iloc[:,4:]
df.head()
# Mental note: The 'related' column uses a strange 1/2 notation
# 1 = related to a disaster, 2 = not related to a disaster instead of the more consistent
# notation of 0/1 for not related vs. related. 
# Inconsequential for this exercise anyway. 
```

### 2. Write a tokenization and lemmatization function to process your text data

```python
def tokenize(text):
    #Convert to lower case, remove non-alphanumeric chars
    text = re.sub(r"[^a-zA-Z0-9]", " ", text.lower())
    stop_words = stopwords.words('english')
    # tokenize text
    tokens = word_tokenize(text)
    # initiate lemmatizer
    lemmatizer = WordNetLemmatizer()
    #lemmatizing and eliminating stop words
    words_lemmed = [WordNetLemmatizer().lemmatize(w) for w in tokens if w not in stop_words]
   
    return words_lemmed
```

### 3. Build a machine learning pipeline
This machine pipeline should take in the `message` column as input and output classification results on the other 36 categories in the dataset. You may find the [MultiOutputClassifier](http://scikit-learn.org/stable/modules/generated/sklearn.multioutput.MultiOutputClassifier.html) helpful for predicting multiple target variables.

```python
pipeline =  Pipeline([
    ('vect', CountVectorizer(tokenizer=tokenize)), # Initial vectorization
    ('tfidf', TfidfTransformer()),                 # TF/IDF to highlight term frequency based significance
    ('clf', multioutput.MultiOutputClassifier(RandomForestClassifier()))  # Wrap RandomForestClassifier in MultiOutputClassifier
    ])
```

### 4. Train pipeline
- Split data into train and test sets
- Train pipeline

```python
# Split into train/test data, reserve 33% for testing 
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 44, test_size=0.33)
# train classifier
pipeline.fit(X_train, y_train)
```

### 5. Test your model
Report the f1 score, precision and recall for each output category of the dataset. You can do this by iterating through the columns and calling sklearn's `classification_report` on each.

```python
# Run prediction on test data
y_pred = pipeline.predict(X_test)
```

```python
# Get results and add them to a dataframe to get aggregate f_score, precision, recall
# MANUAL APPROACH: Gratefully acknowledging work done by isakkabir here
# https://github.com/isakkabir/Disaster-Response-ML-Pipeline/blob/master/ML%20Pipeline%20Preparation.ipynb

def get_results(y_test, y_pred):
    results = pd.DataFrame(columns=['Category', 'f_score', 'precision', 'recall'])
    num = 0
    for cat in y_test.columns:
        precision, recall, f_score, support = precision_recall_fscore_support(y_test[cat], y_pred[:,num], average='weighted')
        results.set_value(num+1, 'Category', cat)
        results.set_value(num+1, 'f_score', f_score)
        results.set_value(num+1, 'precision', precision)
        results.set_value(num+1, 'recall', recall)
        num += 1
    print('Aggregated f_score:', results['f_score'].mean())
    print('Aggregated precision:', results['precision'].mean())
    print('Aggregated recall:', results['recall'].mean())
    return results
```

```python
results_RF_Raw = get_results(y_test, y_pred)
results_RF_Raw
```

### 6. Improve your model
Use grid search to find better parameters. 

```python
# Get all available parameters
pipeline.get_params()
```

```python
# Initially tried larger grid search, but performance just wasn't acceptable. 
# Selectively commented out params to expedite
parameters = {
    'vect__min_df': [1, 5],
 #   'tfidf__use_idf':[True, False],
    'clf__estimator__max_depth': (25, 100, None),
    'clf__estimator__min_samples_split': (2, 50, 100), 
 #    'clf__estimator__n_estimators': [200]    
    }

cv = GridSearchCV(pipeline, parameters, verbose=2)
```

```python
cv.fit(X_train, y_train)
```

### 7. Test your model
Show the accuracy, precision, and recall of the tuned model.

```python
y_pred = cv.predict(X_test)
results_RF_GridSearch = get_results(y_test, y_pred)
```

### 8. Try improving your model further. Here are a few ideas:
* try other machine learning algorithms
* add other features besides the TF-IDF

```python
# Trying ADABoost instead
from sklearn.feature_extraction.text import TfidfVectorizer
pipelineAdaBoost = Pipeline([
    #('vect', CountVectorizer(tokenizer=tokenize)), # Initial vectorization
    ('vect', TfidfVectorizer(tokenizer=tokenize)),                 # TF/IDF to highlight term frequency based significance
    ('clf', multioutput.MultiOutputClassifier(AdaBoostClassifier()))  # Wrap AdaBoostClassifier in MultiOutputClassifier
    ]) 
```

```python
pipelineAdaBoost.get_params()
```

```python
# train AdaBoost classifier
AdaBoostModel = pipelineAdaBoost.fit(X_train, y_train)
# Run prediction on test data
#y_pred = pipelineAdaBoost.predict(X_test)
y_pred = AdaBoostModel.predict(X_test)
# Get results
results_AdaBoost = get_results(y_test, y_pred)
```

AdaBoost without any optimization outperforms the GridSearch version of RandomForest, so this is definitely a promising path to pursue for further optimization. 

```python
# Set up another grid search, this time for AdaBoost pipeline
param_grid_Ada = {
    #'clf__estimator__n_estimators': [100, 200],
    'clf__estimator__learning_rate':[0.1, 0.5, 1]
    }
# create grid search object
cv_ada = GridSearchCV(pipelineAdaBoost, param_grid_Ada, verbose=2)
# Fit on data
cv_ada.fit(X_train, y_train)
```

```python
# Run test data through model and store results
y_pred = cv_ada.predict(X_test)   
results_AdaBoost_GridSearch = get_results(y_test, y_pred)
```
